## Funkcje pomocnicze

```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
def bar_plot(labels, times):
    assert len(labels) == len(times), 'Parameters not valid'
    fig = plt.figure(figsize=(15,10))
    plt.bar(labels, times)
    plt.ylabel("Średni łączny czas oczekiwania filozofa")
    plt.show()
```

## Rozwiązanie w javie

```python
labels = ['Z zagłodzeniem', 'Z arbitrem']
```

10 Filozofów, Czas 5 sekund

```python
bar_plot(labels, [331,2685])
```

10 Filozofów, Czas 10 sekund

```python
bar_plot(labels, [662, 3452])
```
